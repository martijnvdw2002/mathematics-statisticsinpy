```python
from sympy import diff
from sympy.plotting import plot3d
from sympy import symbols

x = symbols('x')
f = x**2
dx_f = diff(f)
plot(f)
print('The derivative of the function is:', dx_f)
```

```python
#partial derivatives
#f(x,y) = 2*x**3 + 3*y**3

from sympy import symbols
from sympy.plotting import plot3d

x, y = symbols('x y')
f = 2*x**3 + 3*y**3
dx_f = diff(f, x)
dy_f = diff(f, y)
print(dx_f)
print(dy_f)
plot3d(f)
```

```python
x, y = symbols('x y')
f = (x**2 + 1)**3 - 2
dx_f = diff(f, x)
dy_f = diff(f, y)
print(dx_f)
print(dy_f)
plot3d(f)
```

```python
x, y = symbols('x y')
f = x**2 + 1
dx_f = diff(f, x)
dy_f = diff(f, y)
print(dx_f)
print(dy_f)
plot3d(f)
```

```python
x, y = symbols('x y')
z = y**3 - 2
dx_z = diff(z, x)
dy_z = diff(z, y)
print(dx_z)
print(dy_z)
plot3d(z)
```
